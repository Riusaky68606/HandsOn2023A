Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame


```python
from statistics import mode

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('datos.csv')

# Columna analizar
nombre_columna = 'edad' 

# Convertir la columna en un arreglo NumPy para calcular la moda
columna_np = np.array(df[nombre_columna])

# Calcular la media, mediana y moda
media_columna = df[nombre_columna].mean()
mediana_columna = df[nombre_columna].median()
moda_columna = mode(columna_np)

# Imprimir los resultados
print("Media:", media_columna)
print("Mediana:", mediana_columna)
print("Moda:", moda_columna)
```

12. Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone

```python
# Obtener el promedio de los parámetros por mes
average_params_per_month = df.groupby('Month')[['TempC', 'Wind', 'Solar.R', 'Ozone']].mean()

# Mostrar el resultado
print(average_params_per_month)
```

13. What values did you obtain regarding ozone and solar radiation attributes? Why? Make the necessary change so that you get the average of the registered values.

```python
# Especificar las funciones de agregación para cada columna
aggregations = {
    'TempC': 'mean',
    'Wind': 'mean',
    'Solar.R': 'mean',
    'Ozone': 'mean'
}

# Obtener el promedio de los parámetros por mes
average_params_per_month = df.groupby('Month').agg(aggregations)

# Mostrar el resultado
print(average_params_per_month)
```
