2. Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column.


```r
# Establecer una semilla para reproducibilidad
set.seed(123)

# Crear una matriz bidimensional de 3 filas y 4 columnas con valores enteros aleatorios
matriz_2d <- matrix(data = sample.int(10, 12, replace = TRUE), nrow = 3, ncol = 4)
print(matriz_2d)

# Calcular la suma de cada fila
suma_filas <- rowSums(matriz_2d)
print(suma_filas)

# Calcular la suma de cada columna
suma_columnas <- colSums(matriz_2d)
print(suma_columnas)
```

4. Load the package dplyr and save the data set in a table data frame format.

```r
library(dplyr)
library(tibble)

df_table <- as_tibble(airquality)
```

5. Select the days in May with a temperature above 70 Fahrenheit.

```r


# Filtrar los días de mayo con temperatura superior a 70 Fahrenheit
may_above_70 <- airquality %>%
  filter(Month == 5 & Temp > 70)

# Verificar los resultados
print(may_above_70)
```

6. Create a new attribute TempC which represents the temperature values in Celsius

```r


# Crear el nuevo atributo TempC en el data frame "df"
df_table <- df_table %>%
  mutate(TempC = (Temp - 32) * (5/9))

# Verificar los resultados
head(df_table)

```

7. Inspect which were the 30 hottest days.

```r

# Ordenar el data frame por la columna de temperatura descendente
hottest_days <- df_table %>% 
  arrange(desc(Temp)) %>% 
  head(30)

# Verificar los resultados
print(hottest_days)
```

8. Inspect which were the hottest days, but also with the highest ozone values.

```r

# Ordenar el data frame por temperatura descendente y ozono descendente
hottest_ozone_days <- df_table %>%
  arrange(desc(Temp), desc(Ozone)) %>%
  head()

# Verificar los resultados
print(hottest_ozone_days)
```

9. Inspect the number of days for which there was a register for each month.

```r

# Contar el número de días con registro para cada mes
days_per_month <- df_table %>%
  group_by(Month) %>%
  count()

# Verificar los resultados
print(days_per_month)
```

10. For each month, obtain the minimum and the maximum temperature registered in Celsius.

```r
# Obtener la temperatura mínima y máxima por mes en Celsius
temperature_stats <- df_table %>%
  mutate(TempC = (Temp - 32) * (5/9)) %>%
  group_by(Month) %>%
  summarise(Min_Temp = min(TempC), Max_Temp = max(TempC))

# Verificar los resultados
print(temperature_stats)
```

11. Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone. 

```r
# Group the data frame by month and calculate the average for Temp, Wind, Solar.R, and Ozone
averages <- aggregate(cbind(Temp, Wind, Solar.R, Ozone) ~ Month, data = df_table, FUN = mean)

# Print the averages
print(averages)
```

12. What values did you obtain regarding ozone and solar radiation attributes? Why? Make the necessary change so that you get the average of the registered values.

```r
# Calcular el promedio general de las columnas 'Temp', 'Wind', 'Solar.R' y 'Ozone'
overall_averages <- colMeans(averages[, c('Temp', 'Wind', 'Solar.R', 'Ozone')], na.rm = TRUE)

# Mostrar los promedios generales
print(overall_averages)
```
